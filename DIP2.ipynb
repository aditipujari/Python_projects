# OPERATIONS ON IMAGE

# Reading Image

```python
from PIL import Image

img = Image.open('D:/Image1.jpeg')
print("\n Default size: ",img.size)
img.show()


```

# Printing Hstogram

```python
print(img.histogram())
print("\n\nLength of the list is: ",len(img.histogram()))
```

# Gray Image and Histogram

```python



from PIL import ImageOps as iops

gray = iops.grayscale(img)
print("\nSize of gray scale image is: ",gray.size)
hist = gray.histogram()
print("\n",hist)
print("\nLength of the list is:  ",len(hist))
```

# Binary Image and Histogram

```python
binary = img.convert(mode = '1', dither = Image.NONE) 
print("\nSize of binary image is: ",binary.size)
binhist = binary.histogram()
print("\n",binhist)
print("\nLength of the list is:  ",len(binhist))
```
